<h2>Столбец с максимальным средним значением после нормировки признаков через
стандартное отклонение?</h2>
<p>Сделаем нормировку числовых признаков через СКО:</p>

```python
import math
numeric = ["price", "bedrooms", "bathrooms", "sqft_living", "sqft_lot", "floors", "view", "condition", "sqft_above", "sqft_basement", "yr_built", "yr_renovated"]
for column in numeric:
    data[column] = (data[column] - data[column].mean())/(math.sqrt(data[column].var()))
```

теперь найдем столбец с наибольшим средним:

```python
max_mean = -math.inf
col_name = ""

for column in numeric:
    if (data[column].mean() > max_mean):
        max_mean = data[column].mean()
        col_name = column
print(col_name, max_mean)
```

<p>Столбец с максимальным средним значением после нормировки признаков через
стандартное отклонение - yr_built (3.270668759979423e-15)</p>
<p>Преобразуем тектовые данные в числовые (сейчас это наиболее подходящий момент):</p>

```python
text = ['date', 'street', 'city', 'statezip', 'country']
for column in text:
    cur = set(data[column])
    i = 0
    for item in sorted(cur): #лексикографический порядок
        data[column] = data[column].replace(item, i)
        i = i + 1
```

train_test_split с параметрами test_size = 0.3, random_state = 42?</h2>

<h2>Между какими признаками наблюдается линейная зависимость (корреляция)?</h2>
<p>Выполним код:</p>

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["column_name"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
```

<p>Выведем ту часть таблицы, где vif >= 10:</p>

```python
vif.loc[vif['VIF'] >= 10.0]
```

<p>значение vif, близкое к бесконечности, означает почти линейную зависимость у этиътрех параметров</p>

<h2>Сколько признаков достаточно для объяснения 90% дисперсии после применения
метода PCA?</h2>
<p>выполним код:</p>

```python
from sklearn import preprocessing
stand_X = pd.DataFrame(preprocessing.scale(x_train), columns = x_train.columns)

from sklearn.decomposition import PCA
for i in range(1, len(stand_X.columns)):
    pca = PCA(n_components = i)
    pca.fit(stand_X)
    print(i, sum(pca.explained_variance_ratio_))
```

<p>как мы видим, для дисперсии в 90% достаточно 11 признаков</p>

<h2>Какой признак вносит наибольший вклад в первую компоненту?</h2>
<p>выполним код:</p>

```python
first_component = pca.components_[0]
max_element = first_component[0]
max_index = 0

for i in range(0, len(first_component)):
    if (abs(first_component[i]) > max_element):
        max_element = abs(first_component[i])
        max_index = i
print(max_index, max_element)
```

<p>отсюда видно, что наибольший вклад в 1-ю компоненту вносит 4-й столбец (индексация с 0), посмотрим, что это:</P>

<h2>Построить двухмерное представление данных с помощью алгоритма t-SNE.  
    На сколько кластеров визуально на ваш взгляд разделяется выборка?<\n2>

```python
lx = pca.transform(stand_X)

from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=0)
tsne_res = tsne.fit_transform(lx)
tsne_data = pd.DataFrame({'X':tsne_res[:,0],
                        'Y':tsne_res[:,1],
                        'real_ans':y_train})

import matplotlib.pyplot as plt
import seaborn as sns
#%matplotlib inline
plt.figure(figsize=(7, 7))
sns.scatterplot(x="X", y="Y",
              data=tsne_data);
```

<p>На мой взгляд, визуально выборку можно разделить на 7 - 9 кластеров, так как четкого разделения тут не наблюдается.<\p>
